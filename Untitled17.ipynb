## **0. Описание задачи "Прогнозирование инсульта"**

Инсульт является одной из главных причин смертности и инвалидности, как в мире, так и в России. За последние 20 лет инсульт становится одним из наиболее распространенных заболеваний людей среднего и пожилого возраста.

Инсульт — это состояние, развивающееся в результате нарушения мозгового кровообращения, которое ведет к значительному ослаблению или прекращению поступления крови к части мозга. Не получая в достаточном количестве кислород и необходимые питательные вещества, уже через несколько минут начинается прогрессирующий процесс гибели клеток головного мозга. Пораженные участки мозга теряют свои функциональные способности, что негативно сказывается на работе органов, которые контролируются этими участками.

Этот набор данных используется для прогнозирования вероятности инсульта у пациента на основе входных параметров, таких как пол, возраст, различные заболевания и статус курения.

Информация об атрибутах:

id: уникальный идентификатор

gender: "Мужской", "Женский" или "Другой"

age: возраст пациента

hypertension: 0 - у пациента нет гипертонии, 1 - у пациента гипертония

heart_disease: 0 - у пациента нет сердечных заболеваний, 1 - у пациента есть заболевание сердца

ever_married: был ли пациент женат - "Нет" или "Да"

work_type: тип работы-"частный", "самозанятый", "другое"

Residence_type: тип проживания- «сельский» или «городской»

avg_glucose_level: средний уровень глюкозы в крови

bmi: индекс массы тела

smoking_status: статус курения- «ранее курил», «никогда не курил», «курит» или «неизвестно»

stroke: 1 - у пациента был инсульт, 0 - нет

## **3. Обработка пропущенных значений.**

Посмотрим, сколько пропущенных значений в каждом столбце матрицы:

```python
data.isnull().sum()
```

Пропущенные значения есть у одного признака "bmi" и их относительно мало, поэтому их можно просто удалить или можно на значение медианы. Воспользуемся вторым способом.

```python
data['bmi'] = data['bmi'].fillna(data['bmi'].median())
data.isnull().sum()
```

Также у признака "smoking_status" есть тип "Unknown", заменим его на Nan.

```python
for i in range(data.shape[0]):
  if data['smoking_status'][i] =="Unknown":
     data['smoking_status'][i]=np.NaN
print(data.isna().sum())
```

Для катекориальных значений будем использовать моду.

```python
data['smoking_status'].fillna(data['smoking_status'].mode().iloc[0], inplace=True)
data.isna().sum()
```

## **4. Обработка категориальных признаков**

Заменим все категориальные признаки на числовые.

Бинарный признак "gender" заменим на числовой. Теперь 0 - Male, 1 - Female.

```python
data['gender'] = pd.factorize(data['gender'])[0]
data.groupby('gender')['gender'].count()
```

Бинарный признак "ever_married" приводим так же к числовому: 0 - Yes, 1 - No.

```python
print(data.groupby('ever_married')['ever_married'].count())
data['ever_married'] = pd.factorize(data['ever_married'])[0]
data.groupby('ever_married')['ever_married'].count()
```

К категориальному (небинарному) признаку "work_type" применим метод бинаризации (one-hot encoding), который заключается в следующем.

Этот признак принимает 5 значений: Govt_job, Never_worked, Private, Self-employed, children.

Вместо признака "work_type" будем использовать 5 новых признаков (dummy-признаков, dummy - фиктивный), которые так и назовем Govt_job, Never_worked, Private, Self-employed, children.
При этом:

*   если признак "work_type" принимает значение "Govt_job", то признак "Govt_job" равен 1, а все остальные 0;
*   и тд.

```python
Work = pd.get_dummies(data['work_type'], dtype='int')
print(Work.head())
```

Добавим эти dummy-столбцы к таблице и удалим столбец "work_type":

```python
data = pd.concat((data, Work), axis=1)
data = data.drop(['work_type'], axis=1)
data.head()
```

Бинарный признак "Residence_type" приводим так же к числовому: 0 - Urban, 1 - Rural

```python
print(data.groupby('Residence_type')['Residence_type'].count())
data['Residence_type'] = pd.factorize(data['Residence_type'])[0]
data.groupby('Residence_type')['Residence_type'].count()
```

Обрабатываем признак "smoking_status":

```python
smoking = pd.get_dummies(data['smoking_status'],dtype='int')
print(smoking.head())
data = pd.concat((data, smoking), axis=1)
data = data.drop(['smoking_status'], axis=1)
data = data.drop(['Unknown'], axis=1)
```

## **6.Разбиение данных на обучающую и тестовую выборки**

```python
X = data[data.columns[:-1]]
y = data['stroke']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)

N_train, _ = X_train.shape
N_test,  _ = X_test.shape

N_train, N_test
```

## **7.Запуск классификатора k ближайших соседей**

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

k_n = KNeighborsClassifier(n_neighbors=3)#количество соседей = 3
k_n.fit(X_train, y_train)

y_train_predict = k_n.predict(X_train)
y_test_predict = k_n.predict(X_test)

print('err_train=', np.mean(y_train != y_train_predict), 'err_test=', np.mean(y_test != y_test_predict))
print(f'Accuracy Score train : {accuracy_score(y_train, y_train_predict)*100}')
print(f'Accuracy Score test : {accuracy_score(y_test, y_test_predict)*100}')
```

## **8-9.Вычислений ошибок на обучающей и тестовой выборках. Выбор оптимального значение гиперпараметра (к- ближайших соседей)**

```python
from sklearn.model_selection import GridSearchCV
nnb = list(range(1, 30))
knn = KNeighborsClassifier()
grid = GridSearchCV(knn, param_grid = {'n_neighbors': nnb}, cv=10)
grid.fit(X_train, y_train)
best_cv_err = 1 - grid.best_score_
best_n_neighbors = grid.best_estimator_.n_neighbors
print(best_cv_err, best_n_neighbors)
```

```python
knn = KNeighborsClassifier(n_neighbors = best_n_neighbors).fit(X_train, y_train)

err_train = np.mean(y_train != knn.predict(X_train))
err_test  = np.mean(y_test  != knn.predict(X_test))

print('err_train=', err_train, 'err_test=', err_test)
print(f'Accuracy Score train : {accuracy_score(y_train, knn.predict(X_train))*100}')
print(f'Accuracy Score test : {accuracy_score(y_test, knn.predict(X_test))*100}')
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cm = confusion_matrix(y_test, knn.predict(X_test))
disp = ConfusionMatrixDisplay(confusion_matrix = cm, display_labels = knn.classes_)
disp.plot()
plt.show()
```

Вывод: Метод k-ближайших соседей протестирован при различных значениях параметра. Определение класса при k=3 показазало точность классификации почти 96.3% на обучающей выборке и 93.66% на тестовой выборке, при k равным оптимальному значению точность классификации 95.6% на обучающей выборке и 93.74% на тестовой выборке. Согласно методу ближайших соседей, 1197 пациентов из тестовой выборки предположительно не имели инсульта и его действительно не было; 0 пациентов из тестовой выборки предположительно имели инсульт и он действительно был; 0 пациентов из тестовой выборки предположительно имели инсульт, но на самом деле его не было; 80 пациентов предположительно не имели инсульта, но на самом деле он был, что довольно плохо.

## **11.Борьба с несбалансированнными классами**



Взвешивание классов. Метод позволяет учесть дисбаланс между классами в процессе обучения модели. Он основан на идее того, что модель будет штрафовать более сильно за ошибки в классе-меньшинстве, поощряя более точное предсказание.

```python
from sklearn.utils.class_weight import compute_class_weight
from sklearn.ensemble import RandomForestClassifier

# Получение весов классов
classes = np.unique(y_train)
weights = compute_class_weight(class_weight='balanced', classes=classes, y=y_train)
class_weights = dict(zip(classes, weights))

# Создание модели с взвешиванием классов
model = RandomForestClassifier(class_weight=class_weights)

grid_balanced = GridSearchCV(model, cv=5, param_grid = {'max_depth': [3,6,9],
    'max_leaf_nodes': [3,6 ,9]},scoring='roc_auc',verbose = 2)
grid_balanced.fit(X_train, y_train)
```

```python
grid_balanced.best_params_
```

```python
rf_balanced = ensemble.RandomForestClassifier(max_depth = grid_balanced.best_params_['max_depth'],max_leaf_nodes = grid_balanced.best_params_['max_leaf_nodes'])
rf_balanced.fit(X_train, y_train)
```

```python
err_train_balanced = np.mean(y_train != rf_balanced.predict(X_train))
err_test_balanced  = np.mean(y_test  != rf_balanced.predict(X_test))

print(err_train_balanced, err_test_balanced)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cm = confusion_matrix(y_test, rf_balanced.predict(X_test))
disp = ConfusionMatrixDisplay(confusion_matrix = cm, display_labels = rf_balanced.classes_)
disp.plot()
plt.show()
```

Видим, что ошибка ещё снизилась. У 18 пациентов ошибка в предсказании.
